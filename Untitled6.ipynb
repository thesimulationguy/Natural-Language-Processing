```
import re
def remove_pattern(in_t,pat):
  r=re.findall(pat,in_t)
  for i in r:
    in_t=re.sub(i,'',in_t)
  return in_t
```

```
import numpy as np
overall['clean_tweet']=np.vectorize(remove_pattern)(overall['tweet'],"@[\w]*")
```

```
overall['clean_tweet']=overall['clean_tweet'].str.replace("[^a-zA-Z#]"," ")
```

```
overall['clean_tweet']=overall['clean_tweet'].apply(lambda x:' '.join([w for w in x.split() if len(w)>3]))
```

```
tokens=overall['clean_tweet'].apply(lambda x:x.split())
```

```
model=gensim.models.Word2Vec(tokens,size=200,window=5,min_count=2,sg=1,hs=0,negative=10,workers=2,seed=34)
```

```
model.train(tokens,total_examples=len(overall['clean_tweet']),epochs=20)
```

```
def word_vector(tokens,size):
  vec=np.zeros(size).reshape((1,size))
  count=0
  for word in tokens:
    try:
      vec=vec+model[word].reshape((1,size))
      count+=1
    except KeyError:
      continue
  if count!=0:
    vec/=count
  return vec
```

```
word_array=np.zeros((len(tokens),200))

for i in range(len(tokens)):
  word_array[i,:]=word_vector(tokens[i],200)
wordvec=pd.DataFrame(word_array)
wordvec.shape
```

```
from sklearn.metrics import f1_score
def custom_eval(preds,dtrain):
  labels=dtrain.get_label().astype(np.int)
  preds=(preds>=0.3).astype(np.int)
  return[('f1_score',f1_score(labels,preds))]
```

```
params={
    'colsample':0.9,
    'colsample_bytree':0.5,
    'eta':0.1,
    'max_depth':10,
    'min_child_weight':8,
    'objective':'binary:logistic',
    'subsample':0.9
}
```

```
xgb_model=xgb.train(
    params,
    dvalid,
    feval=custom_eval,
    num_boost_round=1000,
    maximize=True,
    evals=[(dvalid,"Validation")],
    early_stopping_rounds=10
)
```
